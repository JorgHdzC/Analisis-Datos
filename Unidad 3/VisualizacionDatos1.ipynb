```python
import pandas as pd
import numpy as np
```

### **Paso 1. Carga, limpieza y transformacion de datos**

```python
# Importar archivo excel apartier dee el reglon 20
df_canada = pd.read_excel('Canada.xlsx',
                          sheet_name= 'Canada by Citizenship',
skiprows = range(20), skipfooter = 2)

df_canada
```

```python
print(df_canada.shape)
```

```python
df_canada.head()
```

```python
#Eliminar columnas
df_canada.drop(['AREA','REG','Type','Coverage'], axis=1, inplace = True)
df_canada.head()
```

```python
#Renombrar columnas con metodo rename

df_canada.rename(columns = {'OdName':'Country', 'AreaName':'Continent',
                           'RegName':'Region'}, inplace = True)
df_canada.head()
```

```python
# Verificar que todos los tipos de columna que sean string

all(isinstance(column,str) for column in df_canada.columns)
```

```python
#Cambiar todas las columnas a tipo string a travez de un mapeo

df_canada.columns = list(map(str, df_canada.columns))
```

```python
# Verificar que todos los tipos de columna que sean string

all(isinstance(column,str) for column in df_canada.columns)
```

```python
#Indice cambiar el indice por el del pais para que sea primero
df_canada.set_index('Country', inplace = True)
df_canada.head()
```

```python
# Crear columna que contenga la suma de todos las peronas por years

df_canada['Total'] = df_canada.sum(axis= 1)
df_canada.head()
```

```python
#Crear una lista de valores desde year 1980 a 2014
years = list(map(str, range(1980, 2014)))
years
```

### **Parte 2 . Visualizacion de datos con matplotlib**

```python
#importar libreria
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
```

```python
df_canada.sort_values(['Total'], ascending= False, axis=0,inplace= True)
df_canada.head(10)
```

```python
df_top5 = df_canada.head()
df_top5
```

```python
# Transponer renglosnes y columnas
df_top5 = df_top5[years].transpose()
df_top5
```

```python
df_top5.index = df_top5.index.map(int)
df_top5
```

### **Graficacion**

```python
#Graficar el top 5 
df_top5.plot( kind = 'area', stacked = False, figsize = (20, 10))
plt.title('Tendencias de inmigracion de los primeros 5 paises')
plt.ylabel('Numero de inmigrantes')
plt.xlabel('Años')
plt.show()
```

```python
#transparencia 
df_top5.plot( kind = 'area', alpha=0.25 
             ,stacked = False, figsize = (20, 10))
plt.title('Tendencias de inmigracion de los primeros 5 paises')
plt.ylabel('Numero de inmigrantes')
plt.xlabel('Años')
plt.show()
```

```python
ax = df_top5.plot(kind = 'area', alpha = 0.35,
                 figsize = (20,10))
ax.set_title('Tendencias de inmigracion de los primeros 5 paises a canada')
ax.set_ylabel('Numero de inmigrantes')
ax.set_xlabel('Años')
```

```python
df_china = df_canada.loc['China', years]
df_china.head()
```

```python
#El eje x se usa para etiquetar y la 
#longitud de la barra corresponde al valor de la variable
df_china.plot(kind = 'bar', figsize=(10,6))
plt.xlabel('Años')
plt.ylabel('Numero de inmigrantes')
plt.title('Inmigrantes china a canada desde 1980 al 2013')
plt.show()
```

```python
df_chile = df_canada.loc['Chile', years]
df_chile.head()
```

```python
#El eje x se usa para etiquetar y la 
#longitud de la barra corresponde al valor de la variable
df_chile.plot(kind = 'barh', figsize=(10,12))
plt.ylabel('Años')
plt.xlabel('Numero de inmigrantes')
plt.title('Inmigrantes chilenos a canada desde 1980 al 2013')
for index, value in enumerate(df_chile):
    label = format(int(value),',')
    plt.annotate(label, xy=(value, index), color='black')
plt.show()
```

```python
df_mexico = df_canada.loc['Mexico', years]
df_mexico.head()
```

```python
#El eje x se usa para etiquetar y la 
#longitud de la barra corresponde al valor de la variable
df_mexico.plot(kind = 'bar', figsize=(10,6))
plt.xlabel('Años')
plt.ylabel('Numero de inmigrantes')
plt.title('Inmigrantes china a canada desde 1980 al 2013')
plt.show()
```

### Grafico circulares

```python
df_continentes = df_canada.groupby('Continent', axis=0).sum()
df_continentes
```

```python
df_continentes['Total'].plot(kind='pie',
                            figsize=(5,6),
                            autopct='%1.1f%%',
                            startangle= 0,
                            shadow= True)
plt.title('inmigracion a canada por continente (1980- 2013)')
plt.show()
```

```python
explode_list = [0.1,0,0,0,0.1,0.1]
df_continentes['Total'].plot(kind='pie',
                            figsize=(15,6),
                            autopct='%1.1f%%',
                            startangle= 0,
                            shadow= True,
                            labels = None,
                            pctdistance = 1.15,
                            explode = explode_list)
plt.title('inmigracion a canada por continente (1980- 2013)')
plt.axis('equal')
plt.legend(labels = df_continentes.index, loc='upper right')
plt.show()
```
